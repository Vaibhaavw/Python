```python
import subprocess

def is_installed(name):
    try:
        devnull = open(os.devnull)
        subprocess.Popen([name], stdout=devnull, stderr=devnull).communicate()
        return True
    except OSError as e:
        if e.errno == os.errno.ENOENT:
            return False
        else:
            # Something else went wrong while trying to run `name`
            raise

print(is_installed("your-software-name"))
```

```python
def flatten_dict(d):
    flat_list = []
    for k, v in d.items():
        if isinstance(v, dict):
            flat_list.extend(flatten_dict(v))
        else:
            flat_list.append(v)
    return flat_list


d = {'a': 1, 'b': {'c': 2, 'd': 3}, 'e': 4}
print(flatten_dict(d))
```

```python
def concatenate_lists(*args):
    result = []
    for lst in args:
        result += lst
    return result

list1 = [1, 2, 3]
list2 = [4, 5, 6]
list3 = [7, 8, 9]
print(concatenate_lists(list1, list2, list3))
```

```python
pip install PyPDF2

from PyPDF2 import PdfFileReader, PdfFileWriter

def append_pdfs(input_pdf1, input_pdf2, output_pdf):
    # Create a PDF file writer object
    writer = PdfFileWriter()

    # Open and read the first PDF
    with open(input_pdf1, 'rb') as file1:
        reader = PdfFileReader(file1)
        for page_num in range(reader.getNumPages()):
            page = reader.getPage(page_num)
            writer.addPage(page)

    # Open and read the second PDF
    with open(input_pdf2, 'rb') as file2:
        reader = PdfFileReader(file2)
        for page_num in range(reader.getNumPages()):
            page = reader.getPage(page_num)
            writer.addPage(page)

    # Write the combined PDFs to a new file
    with open(output_pdf, 'wb') as output_file:
        writer.write(output_file)

# Test the function
input_pdf1 = "/path/to/your/first.pdf"
input_pdf2 = "/path/to/your/second.pdf"
output_pdf = "/path/to/your/output.pdf"
append_pdfs(input_pdf1, input_pdf2, output_pdf)

```

```python
import os

def filter_word_files(directory):
    word_files = []
    for filename in os.listdir(directory):
        if filename.endswith(".docx"):
            word_files.append(filename)
    return word_files
```

```python
import cv2

def play_video(video_path):
    # Create a VideoCapture object
    video = cv2.VideoCapture(video_path)

    while True:
        # Capture frame-by-frame
        ret, frame = video.read()

        # If the frame was successfully read
        if ret:
            # Display the resulting frame
            cv2.imshow('Video', frame)

            # Exit if ESC key is pressed
            if cv2.waitKey(1) & 0xFF == 27:
                break
        else:
            break

    # Release the VideoCapture object
    video.release()

    # Close all OpenCV windows
    cv2.destroyAllWindows()

    
play_video("/path/to/your/video.mp4")

pip install opencv-python
```

```python
import imaplib
import email
from email.header import decode_header

def access_mailbox(username, password, mail_server):
    # connect to the server
    mail = imaplib.IMAP4_SSL(mail_server)

    # authenticate
    mail.login(username, password)

    # select the mailbox you want to delete in
    # if you want SPAM, use "INBOX.SPAM"
    mail.select("inbox")

    # get all mail IDs
    result, data = mail.uid('search', None, "ALL")
    mail_ids = data[0].split()
    latest_email_id = mail_ids[-1]

    # fetch the email body (RFC822) for the given ID
    result, email_data = mail.uid('fetch', latest_email_id, '(BODY[TEXT])')
    raw_email = email_data[0][1].decode("utf-8")
    email_message = email.message_from_string(raw_email)

    return email_message

# usage
username = 'your_username'
password = 'your_password'
mail_server = 'your_mail_server'
print(access_mailbox(username, password, mail_server))
```
